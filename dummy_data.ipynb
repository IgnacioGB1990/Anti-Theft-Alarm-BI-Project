#### Functions

```python
def fake_dni():
    # Generate eight random numbers
    random_numbers = ''.join(random.choices(string.digits, k=8))

    # Generate a random letter
    random_letter = random.choice(string.ascii_letters).upper()

    # Combine the random numbers and random letter
    random_id = random_numbers + random_letter

    return random_id


def fake_housing():
    values = ["House", "Office"]
    weights = [0.8, 0.2]  # 80% probability for "House" and 20% for "Office"
    return random.choices(values, weights=weights)[0]


def fake_housing_type():
    values = ["Bajo", "Intermedio","Ático"]
    weights = [0.10, 0.75,0.15]  # 80% probability for "House" and 20% for "Office"
    return random.choices(values, weights=weights)[0]

```

#### Generates the dummy data

```python
# Import Libraries
from faker import Faker
import datetime
import pandas as pd
import random
import string


fake = Faker('es_ES')

# Define the start and end dates
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2022, 12, 31)


# Generate fake data
data = []
for _ in range(100):
    housing = fake_housing()
    if housing == "House":
        housing_type = fake_housing_type()
    alarms_off = fake.random_int(min=1, max=15)
    false_alarms = fake.random_int(min=1, max=alarms_off)
    incident_alarm = alarms_off - false_alarms
    avg_response_time = random.randint(3, 59)
    DNI = fake_dni()
    age = fake.random_int(min=25, max=60)
    name = fake.first_name()
    surname = fake.last_name()
    date = fake.date_time_between(start_date=start_date, end_date=end_date).strftime("%d-%m-%Y")
    data.append([housing,housing_type,alarms_off,false_alarms,incident_alarm,avg_response_time,DNI,age, name, surname,date,])                                 

# Create DataFrame
df = pd.DataFrame(data, columns=["Housing","Type","Alarms Off","False Alarm","Incident Alarm","AVG R.TIME (sec)",'DNI',"Age",'Name', 'Surname', 'Date',])

df
```
