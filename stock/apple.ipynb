### Import Libraries

```python
import yfinance as yf
import pandas as pd
```

### Get ticker list of SP500 members

```python
# Example of a URL where you can find the constituents of the S&P 500 index
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'


# Reading tables from the URL
data = pd.read_html(url)

# The first table on the page often contains the S&P 500 constituents
df_members = data[0]


ticker_list = df_members['Symbol'].tolist()
ticker_list
```

### Code for Apple that worked

```python

# Get Apple stock data for the past year
apple = yf.Ticker("AAPL")

# Get historical market data for the past year
df = apple.history(period="1y")

df['Stock'] = "Apple"

df
```

```python
columns = df.columns
columns.insert(0,"Date")
```

### Loop Code for all Stocks

```python


for ticker in ticker_list[0:3]:

    stock = yf.Ticker(f"{ticker}")
    new_data = stock.history(period="1mo")
    new_data['Stock'] = f"{ticker}"

    df = pd.concat([df, new_data], ignore_index=False)
    

df

```
