#### Get SP500 tickers from wikipedia (should check to see if it is updated/reliable)

```python
import pandas as pd

# Example of a URL where you can find the constituents of the S&P 500 index
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'


# Reading tables from the URL
data = pd.read_html(url)

# The first table on the page often contains the S&P 500 constituents
df = data[0]
df
```

```python
ticker_list = df['Symbol'].tolist()
ticker_list
```

```python
inicio = datetime(2022,12,31)
fin = datetime(2023,11,9)


try:
    datos = pdr.get_data_yahoo(ticker_list, inicio, fin)
except pdr._utils.RemoteDataError as e:
    print(f"Error fetching data: {e}")
```

```python
datos
```

```python
first_level_columns = datos.loc[:, datos.columns.get_level_values(0) == 'Adj Close']
first_level_columns
```

```python

header_array = datos.columns.get_level_values(0).unique()
header_array
```

```python
for i in header_array:
    header = i
    selected_df = datos.loc[:, datos.columns.get_level_values(0) == header]
    selected_df.to_csv(f"{header}.csv", index=True)

```
