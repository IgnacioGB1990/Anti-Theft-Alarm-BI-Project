#### Function to generate fake DNI in Spanish Format

```python
def fake_dni():
    # Generate eight random numbers
    random_numbers = ''.join(random.choices(string.digits, k=8))

    # Generate a random letter
    random_letter = random.choice(string.ascii_letters).upper()

    # Combine the random numbers and random letter
    random_id = random_numbers + random_letter

    return random_id
```

```python
# Import Libraries
from faker import Faker
import datetime
import pandas as pd
import random
import string


fake = Faker('es_ES')

# Define the start and end dates
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2022, 12, 31)


# Give % weights to values
values = ["Vivienda", "Oficina"]
weights = [0.8, 0.2] 


# Generate fake data
data = []
for _ in range(100):
    DNI = fake_dni()
    housing = random.choices(values, weights=weights)[0]
    age = fake.random_int(min=25, max=60)
    name = fake.first_name()
    surname = fake.last_name()
    alarms_off = fake.random_int(min=1, max=15)
    # salary = fake.random_int(min=30000, max=100000)
    date = fake.date_time_between(start_date=start_date, end_date=end_date).strftime("%d-%m-%Y")
    data.append([DNI,housing,age, name, surname,date,alarms_off])                                 

# Create DataFrame
df = pd.DataFrame(data, columns=['DNI',"Housing","Age",'Name', 'Surname', 'Date',"Alarms Off"])

df
```
